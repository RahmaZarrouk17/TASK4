```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your cleaned DataFrame is named cleaned_data
runs_data = cleaned_data['Runs']

# Convert runs_data to numeric
runs_numeric = pd.to_numeric(runs_data, errors='coerce')

# Calculate mean and median
mean_runs = runs_numeric.mean()
median_runs = runs_numeric.median()

# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=runs_numeric)
plt.xlabel('Runs')
plt.title('Box Plot of Runs')
plt.show()

# Print mean and median
print(f"Mean Runs: {mean_runs:.2f}")
print(f"Median Runs: {median_runs:.2f}")
```
